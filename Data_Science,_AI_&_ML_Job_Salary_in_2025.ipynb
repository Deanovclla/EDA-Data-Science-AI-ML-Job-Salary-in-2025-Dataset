```python
#Mengecek data duplikat di setiap kolom
check_duplicate = df.duplicated().sum()

print(f"jumlah data yang duplikat = {check_duplicate}")
```

```python
#Mengatasi data duplikat
df = df.drop_duplicates()
```

```python
#Mengecek duplikat setelah di atasi
handle_duplikate = df.duplicated().sum()

print(f"jumlah data yang duplikat ={handle_duplikate}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Barplot
sns.barplot(data=df, x='experience_level', y='salary_in_usd', estimator=np.mean, palette='pastel')
plt.title("Rata-rata Gaji berdasarkan Tingkat Pengalaman", fontsize=14)
plt.xlabel("Tingkat Pengalaman")
plt.ylabel("Rata-rata Gaji (USD)")
plt.tight_layout()
plt.show()
```
